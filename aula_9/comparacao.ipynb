```python
import subprocess
import time
import pandas as pd

df = pd.DataFrame(columns=['A','B','C','D','E','F','G'])

lista_program = ['../aula_5/main','../aula_6/main','../busca_local/main']

for i in lista_program:
        for j in range(1,8):
            with open(f'in{j}.txt') as f:
                start = time.perf_counter()
                proc = subprocess.run([i], input=f.read(), text=True, capture_output=True)
                end = time.perf_counter()
                print('Saída:', proc.stdout)
                print('Stderr:', proc.stderr)
                print('Tempo total(s):', end - start)
            f.close()
            

# for i in range(1,8):

#     with open(f'in{i}.txt') as f:

# #         for j in lista_program:
#         start = time.perf_counter()
#         proc = subprocess.run([lista_program[2]], input=f.read(), text=True, capture_output=True)
#         end = time.perf_counter()
#         print('Saída:', proc.stdout)
#         print('Stderr:', proc.stderr)
#         print('Tempo total(s):', end - start)
#     f.close()
```
