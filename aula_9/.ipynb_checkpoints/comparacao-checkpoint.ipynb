```python
import subprocess
import time
import pandas as pd

df = pd.DataFrame(columns=['Saida','Tempo','Algoritmo'])

lista_program = ['../aula_5/main','../aula_6/main','../busca_local/main']
lista_programs=[]
for i in lista_program:
        lista_=[]
        for j in range(1,8):
            with open(f'in{j}.txt') as f:
#                 print(f.read())
                start = time.perf_counter()
                proc = subprocess.run([i], input=f.read(), text=True, capture_output=True)
                end = time.perf_counter()
                print('Saída:', proc.stdout)
                print("value", proc.stdout.split()[-1])
                print('Stderr:', proc.stderr)
                print('Tempo total(s):', end - start)
                dic = {
                    'Tempo':end-start,
                    'Saida':proc.stdout.split()[-1]
                }
                lista_.append(dic)
            f.close()
        lista_programs.append(lista_)
for i in lista_programs:
    for j in i:
        if i == 0:
            alg = 'Heuristica'
        elif i == 1:
            alg = 'Aleatorio'
        else:
            alg = 'Busca Local'
        j['Algoritmo'] = alg
        print(j)
        df = df.append(j, ignore_index = True)

# for i in range(1,8):

#     with open(f'in{i}.txt') as f:

# #         for j in lista_program:
#         start = time.perf_counter()
#         proc = subprocess.run([lista_program[2]], input=f.read(), text=True, capture_output=True)
#         end = time.perf_counter()
#         print('Saída:', proc.stdout)
#         print('Stderr:', proc.stderr)
#         print('Tempo total(s):', end - start)
#     f.close()
```
